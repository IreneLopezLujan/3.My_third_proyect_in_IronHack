```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
from pandas import Series
import numpy as np
import urllib.request as urllib2
import re
import os 
os.chdir("../Data")
# ponemos usert agent  en headers para evitar baneaos
headers = {
    "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/71.0.3578.80 Chrome/71.0.3578.80 Safari/537.36"
}

# La url principal
url = 'https://www.decantalo.com/en/wine/'
# peticion request y parseo de beautifulsoup
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
#Buscamos el contenedor donde se encuentran los nombres de los vinos y extraemos su texto
names_container=soup.find_all("h3", class_="product-name-container")
names=[name.text.strip() for name in names_container]
```

```python
print(names)
```

```python
#Buscamos el contenedor donde se encuentran la descripcion de los vinos y extraemos su texto
desc_container=soup.find_all("p", class_="product-desc")
desc=[name.text.strip() for name in desc_container]
```

```python
print(desc)
```

```python
#unimos en una tupla vinos y descripcion 
zip_names_desc=list(zip(names,desc))
#Ya tenemos dataframe de la primera pagina
page_1 =pd.DataFrame(zip_names_desc, columns=["name","description"])
page_1
```

```python
#queremos extraer de la descripcion el vinedo. Nos damos cuenta que hay un menu los filtros de vinedo con valores unicos.
#Vamos a descargarnos todos los que hay y despues los comparamos con nuestra lista para intentar extraerlo
#en otra columna
links_winery=soup.find_all("ul", class_="col-lg-12 layered_filter_ul")[4].select("label>a")
text_winery=[ x.text for x in links_winery ]
print(text_winery)

```

```python
#lo mismo hacemos con el tipo de uva , encontramos otro menu de enlaces/filtros para hacer un listrado 
links_types_gra=soup.find_all("ul", class_="col-lg-12 layered_filter_ul")[7].select("label>a")
text_types_gra=[ x.text for x in links_types_gra ]
print(text_types_gra)
```

```python
#limpiamos los dos listados de filtros
types_gra_clean=[re.sub(r" ?\([^)]+\)", "", item) for item in text_types_gra]
winery_clean=[re.sub(r" ?\([^)]+\)", "", item) for item in text_winery]

```

```python
#ya tenemos la pagina 1 y ahora vamos a sacar de la 2 a la 200 , cuya url sigue otro patrón 

urls = []

for i in range(2, 200):
    url = "https://www.decantalo.com/en/wine/?p=" + str(i)
    urls.append(url)

all_product = []

for index,url in enumerate(urls):
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")
    products = soup.findAll('div', {"class": "left-block"})
    all_product.append(products)

resultset = []

for products in all_product:
    for product in products:
        product_details = dict()
        product_details['name'] = product.find("h3", class_="product-name-container").text.strip('\n\r\t": ').strip('\n\r\t": ').strip('\n\r\t": ').strip('\n\r\t": ')
        product_details['description'] = product.find("p", class_="product-desc").text.strip('\n\r\t": ').strip('\n\r\t": ').strip('\n\r\t": ')
       
        resultset.append(product_details)

```

```python
#hacemos un flatten
flatt = [a for a in resultset]
```

```python
#lo pasamos a Dataframe
pages_2_200 = pd.DataFrame(flatt)
pages_2_200.head()
```

```python
#abrimos el csv para trabajar
prueba = pd.read_csv("pages_2_200_dataset.csv")
```

```python
#eliminamos la columna unnamed que sale por dfecto al cargar un csv
prueba = prueba.drop('Unnamed: 0', 1)
```

```python
df_vinos=pd.concat([prueba, page_1], axis=0)
df_vinos #ahora tenemos 24 productos mas 
```

```python
#Creamos una columna nueva (llamada vineria)mediante una condicion , si algunos de los valores
#del filtro de vinerias previamente extraido se encuentran en la columna descripción entonces ponme ese valor en la nueva columna vineria
tipes_winery=winery_clean
cond=[df_vinos['description'].str.contains(tipes,case=False) for tipes in tipes_winery]
df_vinos['Winery']=np.select(cond,tipes_winery,default = 'Sin definir')
```

```python
#Lo mismo hacemos con el tipo de uva 
tipes_grapes=types_gra_clean
cond=[df_vinos['description'].str.contains(tipes,case=False) for tipes in tipes_grapes]
df_vinos['Tipes_grapes']=np.select(cond,tipes_grapes,default = 'Sin definir')
```

```python
#Tambien crearemos una columna basandonos en la descripcion de los colores del vino. En este caso la condicion
#será una lista creada por nosostros con 7 valores unicos identificados 
colors=['Red wine', 'White wine', 'Orange wine', 'Rosé wine','Sweet wine', 'Fortified wine', 'Vermouth']
cond=[df_vinos['description'].str.contains(color,case=False) for color in colors]
df_vinos['Colors']=np.select(cond,colors,default = 'Sin definir')
df_vinos
```
